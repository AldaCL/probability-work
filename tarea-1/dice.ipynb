# Dice rolls distribution

## This project shows the distribution of a dice roll

```python
import dice
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import functools
```

```python
rolls = dice.roll_dice(100000)
```

```python
distribution = dice.calculate_distribution(rolls)
```

```python
distribution
```

```python
histogram = dice.get_histogram(rolls)
```

```python
dice_faces= list(distribution.keys())
probabilities = list(distribution.values())
bar_labels = ['red', 'blue', 'pink', 'orange', 'purple', 'cyan']
bar_colors = ['tab:red', 'tab:blue', 'tab:pink', 'tab:orange', 'tab:purple', 'tab:cyan']
```

## Histogram of occurrences

```python
fig_hist, ax_hist = plt.subplots()
ax_hist.hist(rolls, bins=6)
ax_hist.set_ylabel("Occurrences")
ax_hist.set_title("Histogram")
```

## Probability of occurences of each dice face

```python
fig, ax = plt.subplots()
ax.bar(dice_faces, probabilities, label=dice_faces, color=bar_colors)

ax.set_ylabel('Probability')
ax.set_title('Probability of each dice face')
fig.legend(title='Dice face', loc='outside right')

plt.show()
```
